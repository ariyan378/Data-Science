```python
s = input('Enter Your String :')
flag = True;

for i in range (0,len(s)//2):
    if s[i]!=s[len(s)-i-1]:
        print("Is is not palindrome")
        break

if flag:
    print("It is a palindrome Number")
```

```python
s = input('Enter Your String :')
L=[]
temp=''
for i in s :
    if i != ' ':
        temp+=i
    else:
        L.append(temp)
        temp =''


L.append(temp)
print(L)

```

```python
#string in title format without using title
s = input('Enter Your String :')
L=[]
for i in s.split():
    L.append(i[0].upper()+i[1:].lower())

print(' '.join(L))
```
