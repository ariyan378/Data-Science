```python
# Simple Calculator using if-else


#Taking Input From User
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

#Menu Of Operation

print("Choose operation:")
print("1. Addition (+)")
print("2. Subtraction (-)")
print("3. Multiplication (*)")
print("4. Division (/)")

choice = input("Enter your choice (1/2/3/4): ")

if choice == '1':
    result = num1 + num2
    print("Result:", result)

elif choice == '2':
    result = num1 - num2
    print("Result:", result)

elif choice == '3':
    result = num1 * num2
    print("Result:", result)

elif choice == '4':
    if num2 != 0:
        result = num1 / num2
        print("Result:", result)
    else:
        print("Error: Cannot divide by zero.")

else:
    print("Invalid choice!")
```

```python
#Guessing Game

 #generate a random integer number between 1 to 100

import random
jackpot = random.randint(1,100)
attempt = 1
#Ask User their Guess
guess = int(input("Guess Your Number :"))

while guess != jackpot:
    if guess < jackpot:
        print("wrong! Guess Higher")
    else :
        print("Wrong! Guess Lower")

    guess = int(input("Guess Your Number :"))
    attempt +=1
else :
    print(f"Correct Guess at attempt no {attempt}")

```

```python
#Sequence sum
#1/1!+2/2!....nth term

#taking user form input

n = int(input("Enter Your number :"))
result = 0
fact= 1

for i in range(1,n+1):
    fact = fact * i
    result = result + i/fact

print(int(result))
```

```python
#printing all prime number


#taking input from user
low = int(input('Enter lower range :'))

upper = int (input('Enter  higher range :'))

#printing all prime number using j and if stament

for i in range (low , upper+1):
    for j in range(2,i):
        if i %  j == 0:
            break
##if divisible by any of them than it will skip all code
    else:
        print(i)

```
